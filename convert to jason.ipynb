```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_csv('MoHi_cleaned.csv')
df.head()
```

```python
data = {
    'type': 'FeatureCollection',
    'features': []
}
for i, row in df.iterrows():
    d = {
        'type': 'Feature',
        'properties': {
            'name': row['restaurant'],
            'status': row['status'],
            'tables': row['tables'],
            'loan': row['loan'],
            'jobs': row['jobs']
        },
        'geometry': {
            'type': 'Point',
            'coordinates': [float(row['longitude']), float(row['latitude'])]
        }
    }
    data['features'].append(d)
```
